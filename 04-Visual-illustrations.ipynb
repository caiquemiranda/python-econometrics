### Matplotlib package

#### Simple plot

```python
# Import matplotlib and simple example

import matplotlib.pyplot as plt
import numpy as np

plt.plot(np.arange(10))
plt.savefig("data/list.pdf")
```

### Figures and subplots

#### Figures

```python
# Create Figures

fig = plt.figure(figsize=(16, 8))

print(plt.gcf())
## Figure(1600x800)
```

#### Subplots

```python
# Adding subplots

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

fig.savefig("data/subplots.pdf")
```

```python
# Filling subplots with content

from numpy.random import randn

ax1.plot([5, 7, 4, 3, 1])
ax2.hist(randn(100), bins=20, color="r")
ax3.scatter(np.arange(30), np.arange(30) * randn(30))
ax4.plot(randn(40), "k--")

fig.savefig("data/content.pdf")
```

#### Standard creation of plots

```python
# Standard creation

fig, axes = plt.subplots(2, 3, figsize=(16, 8), sharey=True)
axes[1, 1].plot(np.arange(7), color="r")
axes[0, 2].plot(np.arange(10, 0, -1))

fig.savefig("data/standard.pdf")
```

### Plot type and styles

#### Plot types

```python
# Types

fig, ax = plt.subplots(1, 3, figsize=(16, 8))

ax[0].hist([1, 2, 3, 4, 5, 4, 3, 2, 3, 4, 2, 3, 4, 4],
           bins=5, 
           color="yellow")

x = np.arange(0, 10, 0.1)
y = np.sin(x)

ax[1].fill_between(x, y, 0, color="green")
ax[2].scatter(x, y)

fig.savefig("data/types.pdf")
```

#### Adjusting the spacing around subplots

```python
# Adjust spacing

fig, axes = plt.subplots(2, 2, 
                         sharex=True, 
                         sharey=True)

for i in range(2):
    for j in range(2):
        axes[i][j].plot(randn(10))

plt.subplots_adjust(wspace=0, hspace=0)

fig.savefig("data/spacing.pdf")
```

#### Colors, markers and line styles

```python
# Styles

fig, ax = plt.subplots(1, figsize=(15, 6))

ax.plot(randn(10), 
        linestyle="--", 
        color="darkcyan", 
        marker="p")

fig.savefig("data/style.pdf")
```

#### Ticks and labels

```python
# Ticks and labels - default

fig, ax = plt.subplots(1, figsize=(15, 10))

ax.plot(randn(1000).cumsum())

fig.savefig("data/withoutlabels.pdf")
```

#### Ticks and labels

```python
# Set ticks and labels

ax.set_xticks([0, 250, 500, 750, 1000])
ax.set_xlabel("Days", fontsize=20)

ax.set_ylabel("Change", fontsize=20)

ax.set_title("Simulation", fontsize=30)

fig.savefig("data/labels.pdf")
```

#### Legends

```python
# Set legend

fig = plt.figure(figsize=(15, 10))

ax = fig.add_subplot(1, 1, 1)

ax.plot(randn(1000).cumsum(), label="first")
ax.plot(randn(1000).cumsum(), label="second")
ax.plot(randn(1000).cumsum(), label="third")

ax.legend(loc="best", fontsize=20)

fig.savefig("data/legend.pdf")
```

#### Annotations on a subplot

```python
# Annotations

ax.text(400, -30, "here", fontsize=50)

ax.annotate("there",
            fontsize=40,
            xy=(0, 0),
            xytext=(400, 8),
            arrowprops=dict(facecolor="black",shrink=0.05))

ax.set_yticks([-40, -30, -20, -10, 0, 10, 20, 30, 40])

fig.savefig("data/arrow.pdf")
```

```python
# Annotation Lehman

import pandas as pd
from datetime import datetime

date = datetime(2008, 9, 15)
fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(1, 1, 1)

dow = pd.read_csv("data/dji.csv", index_col=0, parse_dates=True)

close = dow["Close"]
close.plot(ax=ax)

ax.annotate("Lehman Bankruptcy",
            fontsize=30,
            xy=(date, close.loc[date] + 400),
            xytext=(date, 22000),
            arrowprops=dict(facecolor="red",
                            shrink=0.03))

ax.set_title("Dow Jones Industrial Average", size=40)

fig.savefig("data/lehman.pdf")
```

#### Drawing on a subplot

```python
# Drawing

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(1, 1, 1)

ax.set_xticks([0, 1, 2, 3, 4, 5])
ax.set_yticks([0, 1, 2, 3, 4, 5])

rectangle = plt.Rectangle((1.5, 1),
                          width=0.8, height=2,
                          color="red", angle=30)
circ = plt.Circle((3, 3),
                  radius=1, color="blue")

ax.add_patch(rectangle)
ax.add_patch(circ)

fig.savefig("data/draw.pdf")
```

#### Best practice: Visual illustrations

```python
# Best practice Step 1

fig, ax = plt.subplots(1, 1, figsize=(16, 8))
```

```python
# Best practice Step 2

x = np.arange(0, 10, 0.1)
y = np.sin(x)

ax.scatter(x, y)
```

```python
# Best practice Step 3

ax.scatter(x, y, 
           color="green", 
           marker="s")
```

```python
# Best practice Step 4

ax.set_title("Sine wave", fontsize=30)

ax.set_xticks([0, 2.5, 5, 7.5, 10])
ax.set_yticks([-1, 0, 1])

ax.set_ylabel("y-value", fontsize=20)
ax.set_xlabel("x-value", fontsize=20)
```

```python
# Best practice Step 5

ax.scatter(x, y, 
           color="green", 
           marker="s", 
           label="Sine")
```

```python
# Best practice Step 6

ax.plot(np.arange(11) / 10, 
        color="blue", 
        linestyle="-",
        label="Linear")

ax.legend(fontsize=20)
```
